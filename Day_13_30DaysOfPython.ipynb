# 🐍 30 Days of Python – Day 13: List Comprehension

## Basic Syntax
```python
[expression for item in iterable]
[expression for item in iterable if condition]
```

````
# Traditional loop
squares = []
for i in range(1, 11):
    squares.append(i**2)
print(squares)

# List comprehension
squares = [i**2 for i in range(1, 11)]
print(squares)
````

## With Conditions

````
evens = [x for x in range(20) if x % 2 == 0]
print(evens)

odds = [x for x in range(20) if x % 2 != 0]
print(odds)

# Conditional expression
labels = ['even' if x % 2 == 0 else 'odd' for x in range(10)]
print(labels)
````

## Nested List Comprehension

````
matrix = [[i * j for j in range(1, 4)] for i in range(1, 4)]
for row in matrix:
    print(row)

# Flatten a matrix
flat = [x for row in matrix for x in row]
print(flat)
````

## Dictionary & Set Comprehensions

````
# Dict comprehension
squares_dict = {x: x**2 for x in range(1, 6)}
print(squares_dict)

# Set comprehension
unique_sq = {x**2 for x in [-2, -1, 0, 1, 2]}
print(unique_sq)
````

## Generator Expressions

````
# Like list comprehension but lazy — doesn't store all in memory
gen = (x**2 for x in range(10))
print(next(gen))
print(next(gen))
print(list(gen))   # remaining elements
````

## 💪 Exercises – Day 13
1. Create a list of the first 10 cube numbers
2. Filter out negative numbers from a list using comprehension
3. Create a dictionary of words and their lengths
4. Flatten a 3x3 matrix using nested comprehension

````
# Solutions
cubes = [x**3 for x in range(1, 11)]
print(cubes)

nums = [-5, -3, 0, 2, 4, -1, 7]
positives = [x for x in nums if x >= 0]
print(positives)

words = ['Python', 'is', 'awesome', 'and', 'powerful']
lengths = {w: len(w) for w in words}
print(lengths)

matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat = [x for row in matrix for x in row]
print(flat)
````
