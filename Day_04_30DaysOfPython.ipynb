# 🐍 30 Days of Python – Day 4: Strings

## Creating Strings

```
letter = 'P'
greeting = 'Hello, World!'
sentence = "Python is a beautiful language"
multiline = '''This is
a multiline
string'''
print(letter, greeting, sentence)
print(multiline)
```

## String Operations

```
s = 'Python'
print(len(s))          # 6
print(s.upper())       # PYTHON
print(s.lower())       # python
print(s.capitalize())  # Python
print(s.count('p', 0, len(s)))  # case-sensitive count
print(s.startswith('Py'))  # True
print(s.endswith('on'))    # True
```

## String Indexing & Slicing

```
s = '30 Days of Python'
print(s[0])        # '3'
print(s[-1])       # 'n'
print(s[3:7])      # 'Days'
print(s[:2])       # '30'
print(s[10:])      # 'Python'
print(s[::-1])     # reverse string
```

## String Methods

```
s = '  hello world  '
print(s.strip())           # remove whitespace
print(s.strip().split())   # split into list
print(','.join(['a','b','c']))  # join list to string

s2 = 'Python is great'
print(s2.replace('great', 'awesome'))
print(s2.find('is'))      # index of substring
print(s2.index('is'))     # same, but raises error if not found
```

## f-Strings (String Formatting)

```
name = 'Asabeneh'
age = 250
country = 'Finland'
print(f'I am {name}. I am {age} years old. I live in {country}.')
print(f'2 + 3 = {2+3}')
print(f'Pi ≈ {3.14159:.2f}')
```
