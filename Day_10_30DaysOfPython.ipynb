# 🐍 30 Days of Python – Day 10: Loops

## while Loop

```
count = 0
while count < 5:
    print(count)
    count += 1

# break and continue
i = 0
while i < 10:
    if i == 5:
        break
    if i % 2 == 0:
        i += 1
        continue
    print(i)
    i += 1
```

## for Loop

```
for i in range(5):
    print(i)

fruits = ['apple', 'banana', 'mango']
for fruit in fruits:
    print(fruit)

# enumerate
for idx, fruit in enumerate(fruits):
    print(idx, fruit)
```

## Nested Loops

```
# Multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f'{i} x {j} = {i*j}')
    print('---')
```

## range()

```
print(list(range(10)))          # 0-9
print(list(range(1, 10)))       # 1-9
print(list(range(0, 20, 2)))    # even numbers
print(list(range(10, 0, -1)))   # countdown
```

## zip() and map()

```
names = ['Alice', 'Bob', 'Carol']
scores = [90, 85, 92]

for name, score in zip(names, scores):
    print(f'{name}: {score}')

doubled = list(map(lambda x: x*2, [1,2,3,4,5]))
print(doubled)
```

## 💪 Exercises – Day 10
1. Print numbers 0–10 using a while loop
2. Print the multiplication table of 7
3. Iterate over a string and count vowels
4. Use `break` to stop a loop when number > 20

```
# 1.
i = 0
while i <= 10:
    print(i, end=' ')
    i += 1
print()

# 2.
for j in range(1, 11):
    print(f'7 x {j} = {7*j}')

# 3.
text = 'I love Python programming'
vowels = 'aeiouAEIOU'
count = sum(1 for ch in text if ch in vowels)
print(f'Vowels: {count}')

# 4.
for n in range(100):
    if n > 20:
        break
    print(n, end=' ')
print()
```
