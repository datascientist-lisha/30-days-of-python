## Creating a Class

```
class Person:
    # Class variable (shared by all instances)
    species = 'Homo sapiens'
    
    def __init__(self, name, age, country):
        # Instance variables
        self.name = name
        self.age = age
        self.country = country
    
    def greet(self):
        return f'Hi, I am {self.name} from {self.country}!'
    
    def __str__(self):
        return f'Person(name={self.name}, age={self.age})'
    
    def __repr__(self):
        return f"Person('{self.name}', {self.age}, '{self.country}')"

alice = Person('Alice', 25, 'Kenya')
print(alice)
print(repr(alice))
print(alice.greet())
print(alice.species)
```

## Inheritance

```
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound
    
    def speak(self):
        return f'{self.name} says {self.sound}!'

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, 'Woof')
        self.breed = breed
    
    def fetch(self):
        return f'{self.name} fetches the ball!'

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, 'Meow')
    
    def purr(self):
        return f'{self.name} purrs...'

dog = Dog('Buddy', 'Labrador')
cat = Cat('Whiskers')

print(dog.speak())
print(dog.fetch())
print(cat.speak())
print(cat.purr())
print(isinstance(dog, Animal))  # True
```

## Encapsulation

```
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance   # private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f'Deposited {amount}. Balance: {self.__balance}'
        raise ValueError('Amount must be positive')
    
    def withdraw(self, amount):
        if amount > self.__balance:
            raise ValueError('Insufficient funds')
        self.__balance -= amount
        return f'Withdrew {amount}. Balance: {self.__balance}'
    
    @property
    def balance(self):
        return self.__balance
    
    def __str__(self):
        return f'Account({self.owner}, KES {self.__balance})'

acc = BankAccount('Alice', 1000)
print(acc)
print(acc.deposit(500))
print(acc.withdraw(200))
print(f'Balance: KES {acc.balance}')
```

## Polymorphism & Class/Static Methods

```
class Shape:
    def area(self):
        raise NotImplementedError
    
    def __str__(self):
        return f'{self.__class__.__name__} with area {self.area():.2f}'

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        import math
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, w, h):
        self.width, self.height = w, h
    def area(self):
        return self.width * self.height

shapes = [Circle(5), Rectangle(4, 6), Circle(3)]
for s in shapes:
    print(s)

class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    
    @classmethod
    def from_fahrenheit(cls, f):
        return cls((f - 32) * 5/9)
    
    @staticmethod
    def is_valid(celsius):
        return celsius >= -273.15
    
    def __str__(self):
        return f'{self.celsius:.1f}°C'

t = Temperature.from_fahrenheit(100)
print(t)
print(Temperature.is_valid(-300))
```

```
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    
    def passed(self, pass_mark=50):
        return self.grade >= pass_mark
    
    def __str__(self):
        status = 'PASS' if self.passed() else 'FAIL'
        return f'{self.name}: {self.grade} ({status})'

students = [Student('Alice', 85), Student('Bob', 42)]
for s in students:
    print(s)
```
