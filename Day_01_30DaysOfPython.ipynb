## What is Python?
Python is a high-level, open-source, interpreted, object-oriented programming language. It is easy to learn and widely used in web development, data science, machine learning, automation, and more.

## Running Python

```
print('Hello, World!')
print('I am enjoying 30 Days of Python Challenge')
```

## Data Types Overview

```
# Check data types
print(type(10))         # int
print(type(3.14))       # float
print(type(1 + 3j))     # complex
print(type('Asabeneh')) # str
print(type([1,2,3]))    # list
print(type({'a':1}))    # dict
print(type({1,2,3}))    # set
print(type((1,2,3)))    # tuple
print(type(True))       # bool
```

## Arithmetic Operations

```
print(2 + 3)   # Addition
print(10 - 4)  # Subtraction
print(3 * 4)   # Multiplication
print(9 / 2)   # Division
print(9 // 2)  # Floor division
print(9 % 2)   # Modulus
print(2 ** 3)  # Exponent
```
