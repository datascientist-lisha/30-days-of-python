# 🐍 30 Days of Python – Day 6: Tuples

## What is a Tuple?
Tuples are ordered, **immutable** collections. Once created, they cannot be changed.

```
empty_tuple = ()
single = (1,)           # needs trailing comma
fruits = ('banana', 'orange', 'mango', 'lemon')
mixed = (1, 'hello', True, 3.14)
print(fruits, mixed)
print(type(fruits))
```

## Accessing Tuples

```
fruits = ('banana', 'orange', 'mango', 'lemon')
print(fruits[0])
print(fruits[-1])
print(fruits[1:3])
print(len(fruits))
```

## Tuple Methods

```
nums = (1, 2, 3, 1, 2, 1)
print(nums.count(1))   # 3
print(nums.index(2))   # 1 (first occurrence)
```

## Converting Between List and Tuple

```
fruits_list = ['banana', 'orange', 'mango']
fruits_tuple = tuple(fruits_list)
print(fruits_tuple)

back_to_list = list(fruits_tuple)
print(back_to_list)
```

## Tuple Unpacking

```
fruits = ('banana', 'orange', 'mango')
a, b, c = fruits
print(a, b, c)

first, *rest = (1, 2, 3, 4, 5)
print(first, rest)
```

## When to Use Tuples vs Lists
- Use **tuples** for data that should not change (coordinates, RGB values, DB records)
- Use **lists** for data that may change (shopping cart, todo items)
