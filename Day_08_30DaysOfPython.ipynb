# 🐍 30 Days of Python – Day 8: Dictionaries

## What is a Dictionary?
Dictionaries store data as **key-value pairs** — like a real-world dictionary (word: definition).

```
empty = {}
person = {
    'first_name': 'Asabeneh',
    'last_name': 'Yetayeh',
    'age': 250,
    'country': 'Finland',
    'skills': ['JavaScript', 'React', 'Python'],
    'is_married': True
}
print(person)
```

## Accessing Values

```
person = {'name': 'Alice', 'age': 25, 'city': 'Nairobi'}
print(person['name'])
print(person.get('age'))
print(person.get('salary', 'Not found'))  # default value
```

## Modifying Dictionaries

```
person = {'name': 'Alice', 'age': 25}
person['email'] = 'alice@example.com'   # add key
person['age'] = 26                       # update value
print(person)

del person['email']
print(person)

person.update({'city': 'Nairobi', 'age': 27})
print(person)
```

## Dictionary Methods

```
d = {'a': 1, 'b': 2, 'c': 3}
print(d.keys())    # dict_keys(['a','b','c'])
print(d.values())  # dict_values([1, 2, 3])
print(d.items())   # dict_items([('a',1), ...])

print(d.pop('b'))  # removes and returns value
print(d)
print('a' in d)    # membership check
```

## Nested Dictionaries

```
users = {
    'john': {'age': 30, 'city': 'Nairobi'},
    'jane': {'age': 25, 'city': 'Mombasa'}
}
print(users['john']['city'])
for name, info in users.items():
    print(f"{name} lives in {info['city']}") 
```
