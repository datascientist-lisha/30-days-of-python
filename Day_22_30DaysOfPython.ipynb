## Basic Scraping

```
import requests
from bs4 import BeautifulSoup

# Scrape a public page
url = 'https://books.toscrape.com'
try:
    response = requests.get(url, timeout=10)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Get title
    print('Title:', soup.title.text)
    
    # Get all book titles
    books = soup.find_all('article', class_='product_pod')
    print(f'\nFound {len(books)} books on page 1:')
    for book in books[:5]:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        print(f'  {title[:40]}: {price}')
except Exception as e:
    print(f'Error: {e}')
```

## Parsing HTML

```
html = '''
<html>
<body>
<h1 class="title">Python Web Scraping</h1>
<ul id="topics">
  <li class="item">requests</li>
  <li class="item">BeautifulSoup</li>
  <li class="item">Scrapy</li>
</ul>
<a href="https://python.org">Python Website</a>
</body>
</html>
'''

soup = BeautifulSoup(html, 'html.parser')

# Various ways to find elements
print(soup.find('h1').text)
print(soup.find('h1', class_='title').text)
print([li.text for li in soup.find_all('li')])
print(soup.find('a')['href'])
print(soup.select('#topics .item'))  # CSS selector
```

## Scraping Tables

```
html_table = '''
<table>
  <tr><th>Name</th><th>Age</th><th>City</th></tr>
  <tr><td>Alice</td><td>25</td><td>Nairobi</td></tr>
  <tr><td>Bob</td><td>30</td><td>Mombasa</td></tr>
</table>
'''
soup = BeautifulSoup(html_table, 'html.parser')
rows = soup.find_all('tr')
for row in rows:
    cols = [td.text for td in row.find_all(['td','th'])]
    print(cols)
```

## Ethical Scraping
- Always check `robots.txt` before scraping
- Add delays between requests (`time.sleep()`)
- Use headers to identify your bot
- Don't overload servers
- Respect Terms of Service

## 💪 Exercises – Day 22
1. Scrape book titles and prices from `books.toscrape.com`
2. Parse an HTML table and extract all rows
3. Extract all links from a webpage
4. Save scraped data to a CSV file

```
import requests
from bs4 import BeautifulSoup
import csv, time

try:
    r = requests.get('https://books.toscrape.com', timeout=10)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    books_data = []
    for book in soup.find_all('article', class_='product_pod'):
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text.strip()
        rating = book.p['class'][1]
        books_data.append({'Title': title, 'Price': price, 'Rating': rating})
    
    # Save to CSV
    with open('/tmp/books.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['Title','Price','Rating'])
        writer.writeheader()
        writer.writerows(books_data)
    
    print(f'Scraped and saved {len(books_data)} books')
    for b in books_data[:3]:
        print(b)
except Exception as e:
    print(f'Error: {e}')
```
