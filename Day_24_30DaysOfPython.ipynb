# 🐍 30 Days of Python – Day 24: Statistics

```
import statistics
import math

data = [4, 8, 15, 16, 23, 42, 4, 8, 15, 16]

print('Mean:', statistics.mean(data))
print('Median:', statistics.median(data))
print('Mode:', statistics.mode(data))
print('Variance:', statistics.variance(data))
print('Std Dev:', statistics.stdev(data))
print('Min:', min(data))
print('Max:', max(data))
print('Range:', max(data) - min(data))
```

## NumPy for Statistics

```
import numpy as np

data = np.array([4, 8, 15, 16, 23, 42, 4, 8, 15, 16])

print('NumPy Stats:')
print('Mean:', np.mean(data))
print('Median:', np.median(data))
print('Std Dev:', np.std(data))
print('Variance:', np.var(data))
print('25th percentile:', np.percentile(data, 25))
print('75th percentile:', np.percentile(data, 75))
print('IQR:', np.percentile(data, 75) - np.percentile(data, 25))
print('Correlation matrix:')
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])
print(np.corrcoef(x, y))
```

## Frequency Distribution

```
from collections import Counter

scores = [85, 92, 78, 85, 90, 78, 85, 92, 88, 78, 90, 85]

freq = Counter(scores)
print('Frequency:', dict(sorted(freq.items())))

# Relative frequency
n = len(scores)
print('Relative freq:')
for score, count in sorted(freq.items()):
    print(f'  {score}: {count/n:.1%}')
```

## Visualizing with Matplotlib

```
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

data = np.random.normal(50, 10, 1000)

# Histogram
axes[0,0].hist(data, bins=30, color='steelblue', edgecolor='white')
axes[0,0].set_title('Histogram')
axes[0,0].set_xlabel('Value')

# Box plot
axes[0,1].boxplot(data)
axes[0,1].set_title('Box Plot')

# Scatter
x = np.random.randn(100)
y = 2*x + np.random.randn(100)*0.5
axes[1,0].scatter(x, y, alpha=0.6, color='coral')
axes[1,0].set_title('Scatter Plot')

# Bar
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 67]
axes[1,1].bar(categories, values, color='teal')
axes[1,1].set_title('Bar Chart')

plt.tight_layout()
plt.savefig('/tmp/stats_plots.png', dpi=100)
plt.show()
print('Plot saved!')
```

## 💪 Exercises – Day 24
1. Find mean, median, mode of a dataset
2. Calculate the standard deviation manually
3. Create a histogram of normally distributed data
4. Find outliers using the IQR method

```
import numpy as np

data = [23, 45, 12, 67, 34, 89, 12, 45, 23, 67, 100, 5, 45]

# Stats
print(f'Mean: {np.mean(data):.2f}')
print(f'Median: {np.median(data):.2f}')

# Manual std
mean = sum(data) / len(data)
variance = sum((x - mean)**2 for x in data) / len(data)
std = variance ** 0.5
print(f'Std Dev (manual): {std:.2f}')
print(f'Std Dev (numpy): {np.std(data):.2f}')

# IQR outlier detection
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = [x for x in data if x < lower or x > upper]
print(f'Q1={q1}, Q3={q3}, IQR={iqr}')
print(f'Outliers: {outliers}')
```
