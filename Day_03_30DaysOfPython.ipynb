# 🐍 30 Days of Python – Day 3: Operators

## Arithmetic Operators

```
print(10 + 3)   # 13
print(10 - 3)   # 7
print(10 * 3)   # 30
print(10 / 3)   # 3.333...
print(10 // 3)  # 3  (floor division)
print(10 % 3)   # 1  (modulus)
print(10 ** 3)  # 1000 (exponent)
```

## Comparison Operators

```
print(3 > 2)    # True
print(3 >= 3)   # True
print(3 < 2)    # False
print(3 <= 2)   # False
print(3 == 3)   # True
print(3 != 2)   # True
```

## Logical Operators

```
print(3 > 2 and 4 > 3)   # True
print(3 > 2 and 3 < 2)   # False
print(3 > 2 or 3 < 2)    # True
print(not 3 > 2)          # False
```

## Assignment Operators

```
x = 10
x += 3;  print(x)  # 13
x -= 3;  print(x)  # 10
x *= 3;  print(x)  # 30
x /= 3;  print(x)  # 10.0
x //= 3; print(x)  # 3.0
x **= 3; print(x)  # 27.0
x %= 7;  print(x)  # 6.0
```

## Bitwise Operators

```
print(3 & 4)   # AND   → 0
print(3 | 4)   # OR    → 7
print(3 ^ 4)   # XOR   → 7
print(~3)      # NOT   → -4
print(3 << 1)  # Left shift  → 6
print(8 >> 1)  # Right shift → 4
```

## 💪 Exercises – Day 3
1. Convert 38°C to Fahrenheit: `(°C × 9/5) + 32`
2. Find the area of a triangle: base=20, height=10
3. Calculate perimeter of a rectangle: length=10, width=5
4. Check if `is_raining=False` and `is_sunny=True` can coexist

```
# Solutions
celsius = 38
fahrenheit = (celsius * 9/5) + 32
print(f'{celsius}°C = {fahrenheit}°F')

base, height = 20, 10
area = 0.5 * base * height
print(f'Triangle area: {area}')

length, width = 10, 5
perimeter = 2 * (length + width)
print(f'Rectangle perimeter: {perimeter}')

is_raining = False
is_sunny = True
print(not is_raining and is_sunny)
```
