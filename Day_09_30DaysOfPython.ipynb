# 🐍 30 Days of Python – Day 9: Conditionals

## if / elif / else

```
age = 20
if age < 18:
    print('Minor')
elif age < 65:
    print('Adult')
else:
    print('Senior')
```

## Nested Conditions

```
num = 0
if num > 0:
    print('Positive')
elif num == 0:
    print('Zero')
else:
    if num >= -10:
        print('Negative small')
    else:
        print('Negative large')
```

## Short-hand (Ternary) Operator

```
x = 10
label = 'Even' if x % 2 == 0 else 'Odd'
print(label)
```

## Conditions with `in`, `not in`, `is`, `is not`

```
fruits = ['apple', 'banana', 'mango']
print('apple' in fruits)
print('grape' not in fruits)

a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)    # True – same object
print(a is c)    # False – different object
print(a == c)    # True – same value
```

## 💪 Exercises – Day 9
1. Check if a number is positive, negative, or zero
2. Grade a score: A(>=90), B(>=80), C(>=70), D(>=60), F(<60)
3. Check if a year is a leap year
4. Determine if a number is even or odd

```
# 1.
num = -5
if num > 0: print('Positive')
elif num == 0: print('Zero')
else: print('Negative')

# 2.
score = 85
if score >= 90: grade = 'A'
elif score >= 80: grade = 'B'
elif score >= 70: grade = 'C'
elif score >= 60: grade = 'D'
else: grade = 'F'
print(f'Grade: {grade}')

# 3.
year = 2024
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
print(f'{year} is {"a leap" if is_leap else "not a leap"} year')

# 4.
n = 42
print('Even' if n % 2 == 0 else 'Odd')
```
