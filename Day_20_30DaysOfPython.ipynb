## pip – Python Package Installer

```
# Common pip commands (run in terminal, not notebook)
# pip install package_name
# pip install package_name==1.2.3
# pip install -r requirements.txt
# pip uninstall package_name
# pip list
# pip show package_name
# pip freeze > requirements.txt
# pip install --upgrade package_name

import subprocess
result = subprocess.run(['pip', 'list', '--format=columns'], capture_output=True, text=True)
print(result.stdout[:1000])  # Show first 1000 chars
```

## Virtual Environments

```
# Commands to create a virtual environment (run in terminal):
# python -m venv venv          # create
# source venv/bin/activate     # activate (Mac/Linux)
# venv\Scripts\activate       # activate (Windows)
# deactivate                   # deactivate

# Check if running in a virtual environment:
import sys
print('Executable:', sys.executable)
print('In venv:', hasattr(sys, 'real_prefix') or 
      (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix))
```

## Popular Python Packages

```
# Data Science
import numpy as np
import pandas as pd

arr = np.array([1, 2, 3, 4, 5])
print('NumPy array:', arr)
print('Mean:', np.mean(arr))
print('Std:', np.std(arr))

df = pd.DataFrame({'Name': ['Alice','Bob'], 'Score': [90, 85]})
print(df)
```

```
# requests – HTTP library
try:
    import requests
    r = requests.get('https://api.github.com')
    print('Status:', r.status_code)
    print('Rate limit:', r.headers.get('X-RateLimit-Limit'))
except ImportError:
    print('Install requests: pip install requests')
except Exception as e:
    print(f'Error: {e}')
```

## 💪 Exercises – Day 20
1. List all installed packages using `pip list`
2. Check which version of `numpy` and `pandas` are installed
3. Create a requirements.txt for a data science project
4. Create a virtual environment (describe the steps)

```
import pkg_resources

# List key packages
key_packages = ['numpy', 'pandas', 'requests', 'matplotlib']
for pkg in key_packages:
    try:
        version = pkg_resources.get_distribution(pkg).version
        print(f'{pkg}: {version}')
    except pkg_resources.DistributionNotFound:
        print(f'{pkg}: not installed')

# Requirements file content
print('''\n# requirements.txt for Data Science project:
numpy>=1.24.0
pandas>=2.0.0
matplotlib>=3.7.0
seaborn>=0.12.0
scikit-learn>=1.3.0
jupyter>=1.0.0
requests>=2.31.0
''')
```
