# 🐍 30 Days of Python – Day 14: Higher Order Functions

## What are Higher Order Functions?
Functions that take other functions as arguments or return functions as output.

## map()

```
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
print(squared)

names = ['alice', 'bob', 'carol']
capitalized = list(map(str.capitalize, names))
print(capitalized)
```

## filter()

```
nums = range(-5, 11)
positives = list(filter(lambda x: x > 0, nums))
print(positives)

words = ['apple', 'banana', 'cherry', 'avocado']
a_words = list(filter(lambda w: w.startswith('a'), words))
print(a_words)
```

## reduce()

```
from functools import reduce

nums = [1, 2, 3, 4, 5]
total = reduce(lambda acc, x: acc + x, nums)
print(total)    # 15

product = reduce(lambda acc, x: acc * x, nums)
print(product)  # 120
```

## Closures

```
def outer(x):
    def inner(y):
        return x + y
    return inner

add5 = outer(5)
add10 = outer(10)
print(add5(3))   # 8
print(add10(3))  # 13
```

## Decorators

```
def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper

@uppercase_decorator
def greet(name):
    return f'hello, {name}'

print(greet('world'))

# Decorator with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello():
    print('Hello!')

say_hello()
```

## 💪 Exercises – Day 14
1. Use `map()` to convert a list of strings to uppercase
2. Use `filter()` to find all even numbers in a list
3. Use `reduce()` to find the maximum number
4. Create a decorator that logs when a function is called

```
from functools import reduce

# 1.
words = ['python', 'is', 'fun']
print(list(map(str.upper, words)))

# 2.
nums = range(1, 21)
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)

# 3.
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
maximum = reduce(lambda a, b: a if a > b else b, numbers)
print(maximum)

# 4.
def logger(func):
    def wrapper(*args, **kwargs):
        print(f'Calling {func.__name__}...')
        result = func(*args, **kwargs)
        print(f'{func.__name__} returned {result}')
        return result
    return wrapper

@logger
def add(a, b):
    return a + b

add(3, 4)
```
