# 🐍 30 Days of Python – Day 7: Sets

## What is a Set?
A set is an **unordered** collection of **unique** items. Duplicates are automatically removed.

```
st = {1, 2, 3}
fruits = {'banana', 'orange', 'mango', 'banana'}  # duplicate removed
print(fruits)
print(type(fruits))

empty_set = set()   # NOT {} — that creates a dict!
print(type(empty_set))
```

## Set Methods

```
st = {1, 2, 3}
st.add(4)
print(st)

st.add(3)    # no duplicate
print(st)

st.update([5, 6, 7])
print(st)

st.remove(7)  # raises KeyError if not found
st.discard(10)  # safe — no error if missing
print(st)

popped = st.pop()   # removes arbitrary element
print(popped, st)
```

## Set Operations

```
A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6, 7}

print('Union:', A | B)               # all elements
print('Intersection:', A & B)         # common elements
print('Difference A-B:', A - B)       # in A but not B
print('Symmetric Diff:', A ^ B)       # in either but not both
print('Is subset:', {3,4} <= A)
print('Is superset:', A >= {1,2})
```
