## CRUD Operations

```
crud_code = '''
import pymongo
from datetime import datetime

client = pymongo.MongoClient('mongodb://localhost:27017/')
db = client['thirty_days_db']
students = db['students']

# CREATE
student1 = {
    'name': 'Alice',
    'age': 22,
    'city': 'Nairobi',
    'skills': ['Python', 'JavaScript'],
    'created_at': datetime.now()
}
result = students.insert_one(student1)
print('Inserted ID:', result.inserted_id)

# Insert many
many = [
    {'name': 'Bob', 'age': 25, 'city': 'Mombasa', 'skills': ['Java']},
    {'name': 'Carol', 'age': 21, 'city': 'Kisumu', 'skills': ['Python', 'R']},
]
students.insert_many(many)

# READ
all_students = students.find()
for s in all_students:
    print(s)

# Find with filter
nairobi = students.find({'city': 'Nairobi'})
python_devs = students.find({'skills': 'Python'})

# Find one
student = students.find_one({'name': 'Alice'})

# UPDATE
students.update_one({'name': 'Alice'}, {'$set': {'age': 23}})
students.update_many({'city': 'Nairobi'}, {'$push': {'skills': 'MongoDB'}})

# DELETE
students.delete_one({'name': 'Bob'})
students.delete_many({'age': {'$lt': 22}})

# Query operators
# $eq, $ne, $gt, $gte, $lt, $lte
# $in, $nin, $and, $or, $not
older = students.find({'age': {'$gt': 22}})
'''
print(crud_code)
```

## MongoDB with Flask

```
flask_mongo = '''
from flask import Flask, jsonify, request
import pymongo
from bson import ObjectId
import json

app = Flask(__name__)
client = pymongo.MongoClient('mongodb://localhost:27017/')
db = client['myapp']

def serialize(doc):
    doc['_id'] = str(doc['_id'])
    return doc

@app.route('/api/users', methods=['GET'])
def get_users():
    users = list(db.users.find())
    return jsonify([serialize(u) for u in users])

@app.route('/api/users', methods=['POST'])
def create_user():
    data = request.get_json()
    result = db.users.insert_one(data)
    data['_id'] = str(result.inserted_id)
    return jsonify(data), 201
'''
print(flask_mongo)
```
