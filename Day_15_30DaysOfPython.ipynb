## try / except / else / finally

```
# Basic try-except
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f'Error: {e}')

# Multiple exceptions
try:
    x = int('abc')
except (ValueError, TypeError) as e:
    print(f'Conversion error: {e}')

# else and finally
try:
    n = int('42')
except ValueError as e:
    print(f'Error: {e}')
else:
    print(f'Converted successfully: {n}')
finally:
    print('This always runs')
```

## Raising Exceptions

```
def check_age(age):
    if age < 0:
        raise ValueError('Age cannot be negative')
    if age > 150:
        raise ValueError('Age seems unrealistic')
    return f'Valid age: {age}'

try:
    print(check_age(25))
    print(check_age(-5))
except ValueError as e:
    print(f'Invalid: {e}')
```

## Custom Exceptions

```
class InsufficientFundsError(Exception):
    def __init__(self, amount, balance):
        self.message = f'Cannot withdraw {amount}. Balance: {balance}'
        super().__init__(self.message)

def withdraw(amount, balance):
    if amount > balance:
        raise InsufficientFundsError(amount, balance)
    return balance - amount

try:
    new_balance = withdraw(500, 200)
except InsufficientFundsError as e:
    print(e)
```

## 💪 Exercises – Day 15
1. Write a safe division function using try-except
2. Handle `KeyError` when accessing a missing dictionary key
3. Create a custom exception `InvalidAgeError`
4. Use `finally` to ensure a 'cleanup' message always prints

```
# Solutions
def safe_div(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return 'Cannot divide by zero'

print(safe_div(10, 2))
print(safe_div(5, 0))

d = {'name': 'Alice'}
try:
    print(d['age'])
except KeyError as e:
    print(f'Key {e} not found')

class InvalidAgeError(Exception):
    pass

def validate_age(age):
    if age < 0 or age > 150:
        raise InvalidAgeError(f'Invalid age: {age}')
    return age

try:
    validate_age(200)
except InvalidAgeError as e:
    print(e)
finally:
    print('Validation complete')
```
