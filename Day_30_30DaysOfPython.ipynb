## 🚀 Capstone Project: Student Grade Manager

```
import json
from datetime import datetime
from statistics import mean, median, stdev

class GradeManager:
    def __init__(self):
        self.students = {}
    
    def add_student(self, name, scores):
        student_id = len(self.students) + 1
        self.students[student_id] = {
            'id': student_id,
            'name': name,
            'scores': scores,
            'added': datetime.now().strftime('%Y-%m-%d')
        }
        return student_id
    
    def get_grade(self, avg):
        if avg >= 90: return 'A'
        elif avg >= 80: return 'B'
        elif avg >= 70: return 'C'
        elif avg >= 60: return 'D'
        else: return 'F'
    
    def student_report(self, student_id):
        s = self.students.get(student_id)
        if not s: return f'Student {student_id} not found'
        scores = s['scores']
        avg = mean(scores)
        return {
            'name': s['name'],
            'scores': scores,
            'average': round(avg, 2),
            'grade': self.get_grade(avg),
            'highest': max(scores),
            'lowest': min(scores),
            'passed': avg >= 50
        }
    
    def class_report(self):
        if not self.students: return 'No students'
        all_avgs = [mean(s['scores']) for s in self.students.values()]
        return {
            'total_students': len(self.students),
            'class_avg': round(mean(all_avgs), 2),
            'class_median': round(median(all_avgs), 2),
            'class_std': round(stdev(all_avgs), 2) if len(all_avgs) > 1 else 0,
            'pass_rate': f"{sum(1 for a in all_avgs if a >= 50) / len(all_avgs):.1%}"
        }
    
    def top_students(self, n=3):
        ranked = sorted(
            self.students.values(),
            key=lambda s: mean(s['scores']),
            reverse=True
        )
        return [(s['name'], round(mean(s['scores']), 1)) for s in ranked[:n]]

# Demo
gm = GradeManager()
gm.add_student('Alice', [88, 92, 85, 90])
gm.add_student('Bob', [72, 68, 75, 70])
gm.add_student('Carol', [95, 98, 92, 96])
gm.add_student('David', [45, 52, 48, 55])
gm.add_student('Eve', [81, 79, 83, 77])

print('=== Student Reports ===')
for sid in range(1, 4):
    r = gm.student_report(sid)
    print(f"\n{r['name']}: Avg={r['average']} Grade={r['grade']} {'PASS' if r['passed'] else 'FAIL'}")

print('\n=== Class Report ===')
for k, v in gm.class_report().items():
    print(f'  {k}: {v}')

print('\n=== Top 3 Students ===')
for i, (name, avg) in enumerate(gm.top_students(), 1):
    print(f'  {i}. {name}: {avg}')
```
