# 🐍 30 Days of Python – Day 23: Virtual Environment

## Why Virtual Environments?
Virtual environments isolate project dependencies — different projects can use different package versions without conflicts.

## Creating & Using Virtual Environments

```
# Commands (run in terminal):
steps = '''
1. Create virtual environment:
   python -m venv myenv

2. Activate it:
   Mac/Linux:  source myenv/bin/activate
   Windows:    myenv\Scripts\activate

3. Your prompt changes to: (myenv) $

4. Install packages:
   pip install requests pandas flask

5. Save dependencies:
   pip freeze > requirements.txt

6. Deactivate:
   deactivate

7. Recreate from requirements:
   python -m venv newenv
   source newenv/bin/activate
   pip install -r requirements.txt
'''
print(steps)
```

## virtualenv vs venv vs conda

```
tools = {
    'venv': {
        'built_in': True,
        'command': 'python -m venv myenv',
        'best_for': 'Simple Python projects'
    },
    'virtualenv': {
        'built_in': False,
        'install': 'pip install virtualenv',
        'command': 'virtualenv myenv',
        'best_for': 'More features, older Python support'
    },
    'conda': {
        'built_in': False,
        'install': 'Install Anaconda/Miniconda',
        'command': 'conda create -n myenv python=3.11',
        'best_for': 'Data science, manages non-Python deps'
    },
    'pipenv': {
        'built_in': False,
        'install': 'pip install pipenv',
        'command': 'pipenv install',
        'best_for': 'Combines pip + venv + Pipfile'
    }
}
for tool, info in tools.items():
    print(f'\n{tool}:')
    for k, v in info.items():
        print(f'  {k}: {v}')
```

## Project Structure Best Practice

```
project_structure = '''
my_project/
├── venv/                  ← virtual environment (don't commit)
├── src/
│   ├── __init__.py
│   └── main.py
├── tests/
│   └── test_main.py
├── .gitignore             ← include 'venv/'
├── requirements.txt       ← dependencies
├── requirements-dev.txt   ← dev-only deps (pytest, black)
└── README.md
'''
print(project_structure)

gitignore_content = '''
# .gitignore
venv/
__pycache__/
*.pyc
.env
.DS_Store
*.egg-info/
dist/
build/
'''
print('.gitignore contents:', gitignore_content)
```

## Environment Variables with python-dotenv

```
# Install: pip install python-dotenv
# Create a .env file:
env_example = '''
# .env file (never commit to git!)
DATABASE_URL=postgresql://user:pass@localhost/mydb
SECRET_KEY=my-super-secret-key-12345
API_KEY=your-api-key-here
DEBUG=True
'''

# Usage in Python:
usage = '''
from dotenv import load_dotenv
import os

load_dotenv()

db_url = os.getenv('DATABASE_URL')
secret = os.getenv('SECRET_KEY')
debug = os.getenv('DEBUG', 'False') == 'True'
'''
print(env_example)
print(usage)
```
