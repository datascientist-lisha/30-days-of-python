# 🐍 30 Days of Python – Day 2: Variables & Built-in Functions

## Variables
Variables store data. Python uses dynamic typing — you don't declare the type.

```
first_name = 'Asabeneh'
last_name = 'Yetayeh'
country = 'Finland'
city = 'Helsinki'
age = 250
is_married = True
skills = ['HTML', 'CSS', 'JS', 'React', 'Python']
person_info = {
    'firstname': 'Asabeneh',
    'lastname': 'Yetayeh',
    'country': 'Finland',
    'city': 'Helsinki'
}
print(first_name, last_name, country, city, age, is_married)
print('Skills:', skills)
print('Person Info:', person_info)
```

## Multiple Assignment

```
a, b, c = 1, 2, 3
print(a, b, c)

x = y = z = 0
print(x, y, z)
```

## Built-in Functions

```
# Common built-in functions
print(len('Python'))          # length of string
print(type(3.14))             # type of value
print(str(100))               # convert to string
print(int('50'))              # convert to int
print(float('3.14'))          # convert to float
print(input)                  # input function (don't call it here)
print(abs(-10))               # absolute value
print(round(3.7))             # round
print(min(3, 1, 2))           # minimum
print(max(3, 1, 2))           # maximum
print(sum([1, 2, 3, 4, 5]))   # sum of list
```
