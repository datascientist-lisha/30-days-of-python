## What is an API?
API (Application Programming Interface) allows software to communicate. REST APIs use HTTP methods: GET, POST, PUT, PATCH, DELETE.

```
import requests

# GET request — public API
url = 'https://api.github.com/users/Asabeneh'
try:
    response = requests.get(url, timeout=10)
    print('Status:', response.status_code)
    data = response.json()
    print('Name:', data.get('name'))
    print('Followers:', data.get('followers'))
    print('Public repos:', data.get('public_repos'))
    print('Location:', data.get('location'))
except Exception as e:
    print(f'Error: {e}')
```

```
# REST Countries API
try:
    r = requests.get('https://restcountries.com/v3.1/name/kenya', timeout=10)
    if r.status_code == 200:
        country = r.json()[0]
        print('Country:', country['name']['common'])
        print('Capital:', country['capital'][0])
        print('Population:', country['population']:,)
        print('Region:', country['region'])
        print('Currency:', list(country['currencies'].keys())[0])
except Exception as e:
    print(f'Error: {e}')
```

```
# Working with API responses
import requests, json

def fetch_countries():
    try:
        r = requests.get('https://restcountries.com/v3.1/region/africa', timeout=15)
        r.raise_for_status()
        countries = r.json()
        
        # Process data
        africa = [{
            'name': c['name']['common'],
            'population': c.get('population', 0),
            'capital': c.get('capital', ['N/A'])[0]
        } for c in countries]
        
        # Sort by population
        africa.sort(key=lambda x: x['population'], reverse=True)
        
        print(f'African countries count: {len(africa)}')
        print('\nTop 5 by population:')
        for c in africa[:5]:
            print(f"  {c['name']}: {c['population']:,} (Capital: {c['capital']})")
        return africa
    except Exception as e:
        return f'Error: {e}'

result = fetch_countries()
```

4. Display the top 10 most-starred Python repos on GitHub

```
import requests

def get_top_python_repos():
    url = 'https://api.github.com/search/repositories'
    params = {
        'q': 'language:python',
        'sort': 'stars',
        'order': 'desc',
        'per_page': 10
    }
    try:
        r = requests.get(url, params=params, timeout=10)
        r.raise_for_status()
        repos = r.json()['items']
        print('Top 10 Python Repos by Stars:')
        for i, repo in enumerate(repos, 1):
            print(f"{i:2}. {repo['full_name']}: ⭐ {repo['stargazers_count']:,}")
    except requests.Timeout:
        print('Request timed out')
    except requests.HTTPError as e:
        print(f'HTTP error: {e}')
    except Exception as e:
        print(f'Error: {e}')

get_top_python_repos()
```
