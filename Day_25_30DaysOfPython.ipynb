# 🐍 30 Days of Python – Day 25: Pandas

```
import pandas as pd
import numpy as np

# Creating DataFrames
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [25, 30, 22, 35, 28],
    'City': ['Nairobi', 'Mombasa', 'Kisumu', 'Nairobi', 'Eldoret'],
    'Score': [88, 92, 79, 85, 91],
    'Passed': [True, True, True, True, True]
}
df = pd.DataFrame(data)
print(df)
print('\nShape:', df.shape)
print('\nInfo:')
df.info()
```

```
# Basic operations
print(df.head(3))
print(df.tail(2))
print(df.describe())
print(df.columns.tolist())
print(df.dtypes)
```

## Indexing & Filtering

```
# Select column(s)
print(df['Name'])
print(df[['Name', 'Score']])

# Filter rows
print(df[df['Age'] > 25])
print(df[df['City'] == 'Nairobi'])
print(df[(df['Score'] > 85) & (df['Age'] < 35)])

# loc and iloc
print(df.loc[0])                # by label
print(df.iloc[0])               # by position
print(df.loc[1:3, ['Name','Score']])
```

## Data Cleaning

```
# Handling missing values
df_missing = pd.DataFrame({
    'A': [1, 2, None, 4, 5],
    'B': ['a', None, 'c', None, 'e'],
    'C': [1.1, 2.2, 3.3, None, 5.5]
})
print(df_missing)
print('\nNull counts:')
print(df_missing.isnull().sum())

df_filled = df_missing.fillna({'A': df_missing['A'].mean(), 'B': 'Unknown', 'C': 0})
print('\nFilled:')
print(df_filled)

df_dropped = df_missing.dropna()
print('\nDropped NaN rows:')
print(df_dropped)
```

## GroupBy & Aggregations

```
df2 = pd.DataFrame({
    'Department': ['Sales','IT','Sales','IT','HR','HR','Sales'],
    'Employee': ['Alice','Bob','Carol','David','Eve','Frank','Grace'],
    'Salary': [50000, 80000, 55000, 90000, 45000, 48000, 52000],
    'Years': [3, 5, 2, 7, 1, 4, 3]
})

# Group by department
grp = df2.groupby('Department')
print(grp['Salary'].mean())
print(grp['Salary'].agg(['mean','min','max','count']))
```

## 💪 Exercises – Day 25
1. Create a DataFrame of 5 students with name, age, score
2. Filter students who scored above 80
3. Add a `grade` column based on score
4. Group by a category and compute mean

```
# Solutions
students = pd.DataFrame({
    'Name': ['Alice','Bob','Carol','David','Eve'],
    'Age': [20, 21, 19, 22, 20],
    'Score': [88, 72, 95, 65, 81]
})

high_scorers = students[students['Score'] > 80]
print(high_scorers)

def get_grade(score):
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    else: return 'D'

students['Grade'] = students['Score'].apply(get_grade)
print(students)

print(students.groupby('Grade')['Score'].mean())
```
