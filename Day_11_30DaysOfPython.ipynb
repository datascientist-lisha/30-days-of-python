# 🐍 30 Days of Python – Day 11: Functions

## Defining Functions

```
def greet(name):
    return f'Hello, {name}!'

print(greet('Asabeneh'))
```

## Parameters & Arguments

```
# Default parameters
def greet(name, greeting='Hello'):
    return f'{greeting}, {name}!'

print(greet('Alice'))
print(greet('Bob', 'Hi'))

# Keyword arguments
def describe(name, age, city='Unknown'):
    print(f'{name} is {age} years old from {city}')

describe(age=25, name='Alice', city='Nairobi')
```

## *args and **kwargs

```
def sum_all(*args):
    total = 0
    for n in args:
        total += n
    return total

print(sum_all(1, 2, 3, 4, 5))

def print_info(**kwargs):
    for k, v in kwargs.items():
        print(f'{k}: {v}')

print_info(name='Alice', age=25, city='Nairobi')
```

## Lambda Functions

```
square = lambda x: x ** 2
add = lambda x, y: x + y
even = lambda x: x % 2 == 0

print(square(5))
print(add(3, 4))
print(list(filter(even, range(10))))
```

## Scope

```
x = 'global'

def func():
    x = 'local'
    print(x)

func()
print(x)

# global keyword
count = 0
def increment():
    global count
    count += 1

increment()
increment()
print(count)
```

## 💪 Exercises – Day 11
1. Write a function that takes two numbers and returns their sum
2. Write a function with default arguments
3. Write a function that returns min, max, and sum of a list
4. Create a lambda that checks if a string is a palindrome

```
# Solutions
def add(a, b):
    return a + b
print(add(5, 3))

def power(base, exp=2):
    return base ** exp
print(power(3))
print(power(2, 10))

def stats(lst):
    return min(lst), max(lst), sum(lst)
print(stats([3, 1, 4, 1, 5, 9]))

is_palindrome = lambda s: s == s[::-1]
print(is_palindrome('racecar'))
print(is_palindrome('hello'))
```
