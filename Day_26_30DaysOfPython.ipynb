# 🐍 30 Days of Python – Day 26: Python Web with Flask

## What is Flask?
Flask is a lightweight Python web framework for building web applications and APIs. Install with: `pip install flask`

```
# Basic Flask app
flask_app_code = '''
from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route('/')
def home():
    return '<h1>Welcome to 30 Days of Python!</h1>'

@app.route('/about')
def about():
    return '<p>Built with Flask</p>'

@app.route('/api/greet/<name>')
def greet(name):
    return jsonify({'message': f'Hello, {name}!', 'status': 'success'})

@app.route('/api/sum', methods=['POST'])
def calculate_sum():
    data = request.get_json()
    numbers = data.get('numbers', [])
    return jsonify({'sum': sum(numbers)})

if __name__ == '__main__':
    app.run(debug=True, port=5000)
'''
print(flask_app_code)
```

## Flask with Templates (Jinja2)

```
# Template example (templates/index.html):
template_code = '''
<!DOCTYPE html>
<html>
<head><title>{{ title }}</title></head>
<body>
  <h1>{{ heading }}</h1>
  <ul>
    {% for item in items %}
      <li>{{ item }}</li>
    {% endfor %}
  </ul>
  {% if user %}
    <p>Welcome, {{ user }}!</p>
  {% else %}
    <p>Please log in.</p>
  {% endif %}
</body>
</html>
'''

flask_with_template = '''
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html',
        title='30 Days of Python',
        heading='Topics Covered',
        items=['Variables', 'Functions', 'OOP', 'Flask'],
        user='Alice'
    )
'''
print(template_code)
print(flask_with_template)
```

## REST API Design

```
rest_api = '''
from flask import Flask, jsonify, request

app = Flask(__name__)

# In-memory store (use a real DB in production)
users = {
    1: {'name': 'Alice', 'email': 'alice@example.com'},
    2: {'name': 'Bob',   'email': 'bob@example.com'}
}

@app.route('/api/users', methods=['GET'])
def get_users():
    return jsonify(list(users.values()))

@app.route('/api/users/<int:user_id>', methods=['GET'])
def get_user(user_id):
    user = users.get(user_id)
    if user:
        return jsonify(user)
    return jsonify({'error': 'User not found'}), 404

@app.route('/api/users', methods=['POST'])
def create_user():
    data = request.get_json()
    new_id = max(users.keys()) + 1
    users[new_id] = data
    return jsonify({'id': new_id, **data}), 201

@app.route('/api/users/<int:user_id>', methods=['PUT'])
def update_user(user_id):
    if user_id not in users:
        return jsonify({'error': 'Not found'}), 404
    users[user_id].update(request.get_json())
    return jsonify(users[user_id])

@app.route('/api/users/<int:user_id>', methods=['DELETE'])
def delete_user(user_id):
    if user_id not in users:
        return jsonify({'error': 'Not found'}), 404
    del users[user_id]
    return jsonify({'message': 'Deleted'})
'''
print(rest_api)
```

## 💪 Exercises – Day 26
1. Write a Flask app with 3 routes: `/`, `/about`, `/contact`
2. Create an API endpoint that returns a list of books
3. Add a POST endpoint that accepts JSON and echoes it back
4. Add error handling for 404 and 500 errors

```
exercise_code = '''
from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route('/')
def home():
    return jsonify({'page': 'Home', 'message': 'Welcome!'})

@app.route('/about')
def about():
    return jsonify({'page': 'About', 'author': 'Your Name'})

@app.route('/contact')
def contact():
    return jsonify({'page': 'Contact', 'email': 'you@example.com'})

books = [
    {'id': 1, 'title': 'Python Crash Course', 'author': 'Eric Matthes'},
    {'id': 2, 'title': 'Fluent Python', 'author': 'Luciano Ramalho'},
]

@app.route('/api/books')
def get_books():
    return jsonify(books)

@app.route('/api/echo', methods=['POST'])
def echo():
    data = request.get_json()
    return jsonify({'echo': data})

@app.errorhandler(404)
def not_found(e):
    return jsonify({'error': 'Not found'}), 404

@app.errorhandler(500)
def server_error(e):
    return jsonify({'error': 'Internal server error'}), 500
'''
print(exercise_code)
print('\nTo run: python app.py then visit http://localhost:5000')
```
