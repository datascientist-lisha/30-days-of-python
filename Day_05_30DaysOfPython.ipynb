# 🐍 30 Days of Python – Day 5: Lists

## Creating Lists

```
fruits = ['banana', 'orange', 'mango', 'lemon']
numbers = [1, 2, 3, 4, 5]
mixed = ['string', 10, True, 3.14]
empty = []
print(fruits, numbers, mixed, empty)
```

## Accessing & Slicing

```
fruits = ['banana', 'orange', 'mango', 'lemon']
print(fruits[0])     # banana
print(fruits[-1])    # lemon
print(fruits[1:3])   # ['orange', 'mango']
print(len(fruits))   # 4
```

## Modifying Lists

```
fruits = ['banana', 'orange', 'mango']
fruits.append('avocado')
print(fruits)

fruits.insert(1, 'grape')
print(fruits)

fruits.remove('banana')
print(fruits)

popped = fruits.pop()
print(popped, fruits)

fruits[0] = 'watermelon'
print(fruits)
```

## List Methods

```
nums = [3, 1, 4, 1, 5, 9, 2, 6]
print(sorted(nums))      # sorted (new list)
nums.sort()
print(nums)              # in-place sort
nums.reverse()
print(nums)
print(nums.count(1))     # count occurrences
print(nums.index(5))     # index of value
nums.extend([10, 11])
print(nums)
nums.clear()
print(nums)
```

## List Unpacking & Copying

```
first, second, *rest = [1, 2, 3, 4, 5]
print(first, second, rest)

original = [1, 2, 3]
copy = original.copy()
copy.append(4)
print(original)   # unchanged
print(copy)
```

3. Sort a mixed list `[0, 9, 7, 5, 2]` ascending and descending
4. Remove the first 3 items from a list

```
# Solutions
my_list = []
for i in ['a', 'b', 'c', 'd', 'e']:
    my_list.append(i)
print(my_list)

print(my_list[len(my_list)//2])   # middle item

nums = [0, 9, 7, 5, 2]
print(sorted(nums))
print(sorted(nums, reverse=True))

del nums[:3]
print(nums)
```
