# 🐍 30 Days of Python – Day 19: File Handling

## Opening Files
Modes: `r` (read), `w` (write), `a` (append), `x` (create), `rb`/`wb` (binary)

```
# Writing a file
with open('/tmp/sample.txt', 'w') as f:
    f.write('Hello, World!\n')
    f.write('Python file handling is easy.\n')
    f.write('30 Days of Python\n')

print('File written successfully')
```

```
# Reading entire file
with open('/tmp/sample.txt', 'r') as f:
    content = f.read()
    print(content)

# Reading line by line
with open('/tmp/sample.txt', 'r') as f:
    for line in f:
        print(line.strip())

# Read into list
with open('/tmp/sample.txt', 'r') as f:
    lines = f.readlines()
    print(lines)
```

```
# Appending to a file
with open('/tmp/sample.txt', 'a') as f:
    f.write('This line was appended.\n')

with open('/tmp/sample.txt', 'r') as f:
    print(f.read())
```

## Working with CSV

```
import csv

# Write CSV
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 25, 'Nairobi'],
    ['Bob', 30, 'Mombasa'],
    ['Carol', 22, 'Kisumu'],
]

with open('/tmp/people.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)

# Read CSV
with open('/tmp/people.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
```

## Working with JSON

```
import json

# Write JSON
person = {
    'name': 'Asabeneh',
    'age': 250,
    'skills': ['Python', 'JavaScript'],
    'is_married': True,
    'address': {'city': 'Helsinki', 'country': 'Finland'}
}

with open('/tmp/person.json', 'w') as f:
    json.dump(person, f, indent=4)

# Read JSON
with open('/tmp/person.json', 'r') as f:
    loaded = json.load(f)
    print(loaded['name'])
    print(loaded['skills'])

# String operations
json_str = json.dumps(person, indent=2)
print(json_str[:100])

back = json.loads(json_str)
print(type(back))
```

## 💪 Exercises – Day 19
1. Write a list of 5 countries to a file (one per line)
2. Read the file and count the number of lines
3. Write student data to CSV and read it back
4. Create and read a JSON config file

```
import csv, json

# 1 & 2
countries = ['Kenya', 'Uganda', 'Tanzania', 'Rwanda', 'Ethiopia']
with open('/tmp/countries.txt', 'w') as f:
    f.write('\n'.join(countries))

with open('/tmp/countries.txt') as f:
    lines = f.readlines()
    print(f'{len(lines)} countries in file')

# 3
students = [('Alice', 90), ('Bob', 85), ('Carol', 92)]
with open('/tmp/students.csv', 'w', newline='') as f:
    w = csv.writer(f)
    w.writerow(['Name', 'Score'])
    w.writerows(students)

with open('/tmp/students.csv') as f:
    for row in csv.DictReader(f):
        print(row)

# 4
config = {'debug': False, 'version': '1.0', 'db': 'sqlite'}
with open('/tmp/config.json', 'w') as f:
    json.dump(config, f, indent=2)

with open('/tmp/config.json') as f:
    print(json.load(f))
```
