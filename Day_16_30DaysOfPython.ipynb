# 🐍 30 Days of Python – Day 16: Python Datetime

```
from datetime import datetime, date, time, timedelta
import time as t

# Current date and time
now = datetime.now()
print(now)
print(f'Year: {now.year}, Month: {now.month}, Day: {now.day}')
print(f'Hour: {now.hour}, Minute: {now.minute}, Second: {now.second}')
```

## Formatting Dates

```
now = datetime.now()
print(now.strftime('%Y-%m-%d'))           # 2024-01-15
print(now.strftime('%d/%m/%Y'))           # 15/01/2024
print(now.strftime('%B %d, %Y'))          # January 15, 2024
print(now.strftime('%A, %B %d %Y'))       # Monday, January 15 2024
print(now.strftime('%I:%M %p'))           # 03:45 PM
```

## Parsing Dates

```
date_str = '2024-01-15'
dt = datetime.strptime(date_str, '%Y-%m-%d')
print(dt)
print(type(dt))

# ISO format
dt2 = datetime.fromisoformat('2024-01-15T10:30:00')
print(dt2)
```

## timedelta – Date Arithmetic

```
today = date.today()
print(today)

tomorrow = today + timedelta(days=1)
print(tomorrow)

last_week = today - timedelta(weeks=1)
print(last_week)

# Difference between two dates
from datetime import date
independence_day = date(1963, 12, 12)  # Kenya independence
days_since = (date.today() - independence_day).days
print(f'Days since Kenya independence: {days_since}')
```

## 💪 Exercises – Day 16
1. Print today's date in the format `DD Month YYYY`
2. Calculate your age in days from your birthday
3. Find what day of the week Jan 1, 2025 was
4. Add 100 days to today and print the result

```
from datetime import date, timedelta

today = date.today()
print(today.strftime('%d %B %Y'))

birthday = date(2000, 1, 1)   # change to your birthday
age_days = (today - birthday).days
print(f'Age in days: {age_days}')

jan1 = date(2025, 1, 1)
print(jan1.strftime('%A'))   # day of week

future = today + timedelta(days=100)
print(future)
```
