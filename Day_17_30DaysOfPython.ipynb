## Comprehensive Exception Handling

```
def read_file(filename):
    try:
        with open(filename, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return f'File {filename!r} not found'
    except PermissionError:
        return f'No permission to read {filename!r}'
    except Exception as e:
        return f'Unexpected error: {type(e).__name__}: {e}'

print(read_file('nonexistent.txt'))
```

## Context Managers (`with` statement)

```
import os

# Writing a file safely
with open('/tmp/test.txt', 'w') as f:
    f.write('Hello, World!\n')
    f.write('Python is great!\n')

# Reading
with open('/tmp/test.txt', 'r') as f:
    content = f.read()
    print(content)

# File is automatically closed after the with block
```

## Custom Exception Classes

```
class AppError(Exception):
    # Base class for app exceptions
    pass

class ValidationError(AppError):
    def __init__(self, field, message):
        self.field = field
        self.message = message
        super().__init__(f"Validation error on '{field}': {message}")

class DatabaseError(AppError):
    pass

def validate_email(email):
    if '@' not in email:
        raise ValidationError('email', 'Must contain @')
    if '.' not in email.split('@')[-1]:
        raise ValidationError('email', 'Invalid domain')
    return True

for email in ['user@example.com', 'invalid-email', 'no-dot@com']:
    try:
        validate_email(email)
        print(f'{email}: Valid')
    except ValidationError as e:
        print(f'{email}: {e}')
```

## assert Statement

```
def divide(a, b):
    assert b != 0, 'Divisor cannot be zero'
    return a / b

try:
    print(divide(10, 2))
    print(divide(10, 0))
except AssertionError as e:
    print(f'AssertionError: {e}')
```
