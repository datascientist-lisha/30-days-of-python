# 🐍 30 Days of Python – Day 12: Modules

## What is a Module?
A module is a Python file containing functions, classes, and variables. It promotes code reuse.

## Importing Modules

```
import math
print(math.pi)
print(math.sqrt(16))
print(math.ceil(4.2))
print(math.floor(4.8))
print(math.log(100, 10))
print(math.factorial(5))
```

```
import random
print(random.random())           # float 0-1
print(random.randint(1, 100))    # int between 1 and 100
items = ['apple', 'banana', 'mango']
print(random.choice(items))
random.shuffle(items)
print(items)
```

```
import os
print(os.getcwd())               # current directory
# os.mkdir('test_folder')        # create directory
# os.listdir()                   # list files
print(os.path.exists('.'))       # check if path exists
print(os.path.join('folder', 'file.txt'))
```

```
import sys
print(sys.version)
print(sys.path[:2])              # first 2 entries in module path
```

```
from datetime import datetime, date, timedelta
now = datetime.now()
print(now)
print(now.year, now.month, now.day)
print(now.strftime('%Y-%m-%d %H:%M'))

today = date.today()
print(today)
future = today + timedelta(days=30)
print(future)
```

```
import statistics
data = [4, 8, 15, 16, 23, 42]
print(statistics.mean(data))
print(statistics.median(data))
print(statistics.stdev(data))
```

## 💪 Exercises – Day 12
1. Use `math` to compute hypotenuse of a right triangle
2. Generate 10 random numbers between 1 and 100
3. Get today's date and add 60 days
4. Find the mean and median of a number list

```
import math, random
from datetime import date, timedelta
import statistics

# 1.
a, b = 3, 4
print(math.sqrt(a**2 + b**2))

# 2.
print([random.randint(1, 100) for _ in range(10)])

# 3.
today = date.today()
print(today + timedelta(days=60))

# 4.
nums = [12, 7, 3, 14, 6, 11, 5, 4, 15, 9]
print(statistics.mean(nums))
print(statistics.median(nums))
```
